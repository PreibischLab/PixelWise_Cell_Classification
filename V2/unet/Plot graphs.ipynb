```python
import re
def process_line(line,string):
    parts = line.split(' ')
    elm = parts[parts.index(string)+1]
    elm = re.findall(r'([+-]?\d+(?:\.\d+)?(?:[eE][+-]\d+)?)', elm)[0]
    
    v = float(elm)
    return v

def process_file(file,string):
    l1 = [] 
    cnt = 0
    with open(file) as fp:
        i = 0
        line = fp.readline()
        started = False
        tmp = []
        while line:
            cnt = cnt +1
            line = fp.readline()
            if string in line:
                started = True
                v = process_line(line,string)
                tmp.append(v)
            else:
                if started:
                    if i > 1000:
                        l1.append(tmp)
                        tmp = []
        #                 print('restart')
                        started = False
                        i = 0
                    else:
                        i = i+1
        if len(l1)==0:
            l1.append(tmp)
        print(cnt)
        return l1
        
```

```python
strings = ['mean_io_u:','mean_io_u_1:','mean_io_u_1:']
all_history = [process_file(f,strings[i]) for i,f in enumerate(LOGS)]
```
