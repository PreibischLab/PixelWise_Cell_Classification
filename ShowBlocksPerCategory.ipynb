```python
def getPath(x):
    return 'data/raw/image/'+str(x)+"/"
def show(cat,n):
    print(cat)
    path = getPath(cat)
    allImages = os.listdir(path)
    print(len(allImages))
    
    plt.figure(figsize=(50,50))
    for i in range(15):
        path_example = path + allImages[i]
        img = cv2.imread(path_example, cv2.IMREAD_COLOR)
        plt.subplot(5,3,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(img, cmap=plt.cm.binary)
    plt.show()
    
```
